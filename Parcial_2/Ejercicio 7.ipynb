```python
import numpy as np
import matplotlib.pyplot as plt

def f(x, y):
    r = np.sqrt(-(x**2) - (y**2)+1)
    if r <= 1:
        return r
    else:
        return 0

def calcular_volumen_semi_esfera(R, n):
    delta = 2 * R / n
    volume = 0.0
    for i in range(n):
        for j in range(n):
            x = -R + i * delta
            y = -R + j * delta
            avg_value = (f(x, y) + f(x + delta, y) + f(x, y + delta) + f(x + delta, y + delta)) / 4
            volume += avg_value * delta**2
    return volume

# Parámetros
R = 1.0
n = 100

# Crear una grilla para la visualización
x = np.linspace(-R, R, 100)
y = np.linspace(-R, R, 100)
X, Y = np.meshgrid(x, y)
Z = np.zeros_like(X)

for i in range(Z.shape[0]):
    for j in range(Z.shape[1]):
        Z[i, j] = f(X[i, j], Y[i, j])

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z)
ax.set_xlabel('Eje X')
ax.set_ylabel('Eje Y')
ax.set_zlabel('f(x, y)')
ax.set_title('Semiesfera y Grilla')
plt.show()

# Calcular el volumen aproximado
volumen_aproximado = calcular_volumen_semi_esfera(R, n)
print("Volumen aproximado de la semiesfera:", volumen_aproximado)
```
